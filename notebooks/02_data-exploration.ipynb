```python
import pandas as pd
```

```python

def add_features(df: pd.DataFrame) -> pd.DataFrame:
    df["word_count"] = df["text"].str.split().str.len()
    df["exclamation_count"] = df["text"].str.count(r"!")
    df["has_url"] = df["text"].str.contains(
        r"http|www|\.com|\.co\.uk", case=False, regex=True
    ).astype(int)
    df["has_money"] = df["text"].str.contains(
        r"\£|\$|\€|free|win|cash|prize|earn|claim|awarded",
        case=False, regex=True
    ).astype(int)
    df["has_phone"] = df["text"].str.contains(
        r"\b0\d{9,10}\b|\b\+\d{10,13}\b", regex=True
    ).astype(int)
    return df
```

```python
df = add_features(df)
```

```python
import re
```

```python
def _clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", " url ", text)
    text = re.sub(r"\£|\$|\€", " money ", text)
    text = re.sub(r"\b\d+\b", " num ", text)
    text = re.sub(r"[^\w\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```
